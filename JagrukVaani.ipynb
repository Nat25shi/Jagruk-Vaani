```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, BitsAndBytesConfig

# Configuration for efficient loading (4-bit quantization)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Load the model and tokenizer
model_id = "mistralai/Mistral-7B-Instruct-v0.2" # A strong, widely-used open-source model
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

# Create the text generation pipeline
# This wraps the model and tokenizer into one easy-to-use function
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=500, # Max length of the response
    do_sample=True,
    temperature=0.7,
    top_k=50
)
print("Mistral 7B Model Loaded Successfully.")
```

```python
def lookup_related_context(complaint_text, locality):
    related_issues = []
    text_lower = complaint_text.lower()

    # Expanded keywords for better matching
    keywords = {
        "water": ["paani", "water", "leak", "pipe", "नल", "leakage", "tonty", "line"],
        "road": ["road", "gaddha", "pothole", "sadak", "सड़क", "tooti"],
        "drainage": ["drain", "gutter", "drainage", "sewage", "नाला", "ganda"],
        "electricity": ["light", "bijli", "power", "pole", "खंबा", "current"]
    }

    # Determine category
    current_category = "general"
    for category, words in keywords.items():
        if any(word in text_lower for word in words):
            current_category = category
            break

    # Search past data
    for issue in PAST_COMPLAINTS_DATA:
        # Crucial: Exact locality match
        if issue["geo_cluster"] == locality:
            # Check if previous issues match the same category keywords
            desc_lower = issue["description"].lower()
            if any(word in desc_lower for word in keywords.get(current_category, [])):
                related_issues.append(f"- {issue['description']}")

    if len(related_issues) >= 2:
        return "ALERT: Multiple similar issues detected in this area recently. This indicates a systemic failure."
    elif len(related_issues) == 1:
        return "One similar issue reported recently."
    else:
        return "No recent related issues."
```

```python
def generate_master_prompt(user_complaint, context_data, locality):
    """
    Final Master Prompt for Jagruk Vaani.
    Combines Role, Context, Task, and Constraints for Mistral 7B.
    """
    system_instruction = f"""<s>[INST]
    ROLE: You are 'Jagruk Vaani', the AI Municipal Officer for Gurugram, Haryana.

    CONTEXT FOR {locality}:
    {context_data}

    TASK:
    1. Analyze the citizen's 'Input Complaint'.
    2. Cross-reference it with the 'CONTEXT' provided above.
    3. Generate a structured Work Order in JSON format.

    STRICT LOGIC RULES:
    - IF context contains 'ALERT', SET "Priority": "Critical".
    - IF context contains 'No recent issues', SET "Priority": "Medium".
    - Categorize into: PWD_Pothole, WTR_Leak, SNT_Drainage, or ELE_Streetlight.

    OUTPUT FORMAT:
    {{
      "ActionCode": "Category_Name",
      "Priority": "Critical/Medium",
      "Justification": "Short reason based on history"
    }}

    Input Complaint: "{user_complaint}"
    Output: [/INST]"""
    return system_instruction
```

```python
def process_citizen_complaint(complaint_text, locality):
    # 1. Get the context from your lookup function
    context = lookup_related_context(complaint_text, locality)

    # 2. Create the prompt for Mistral
    full_prompt = generate_master_prompt(complaint_text, context, locality)

    # 3. Ask Mistral 7B
    output = pipe(
        full_prompt,
        max_new_tokens=150,
        do_sample=True,
        temperature=0.1,
        return_full_text=False
    )

    # 4. Extract the raw text
    raw_text = output[0]['generated_text']

    # 5. CLEANING LOGIC: This stops the AI from repeating itself
    # It finds the first closing bracket '}' and ignores everything after it.
    if '}' in raw_text:
        clean_json = raw_text.split('}')[0] + '}'
    else:
        clean_json = raw_text # Fallback if JSON is malformed

    return clean_json

print("Process Complaint Function Updated with Cleaning Logic.")
```
